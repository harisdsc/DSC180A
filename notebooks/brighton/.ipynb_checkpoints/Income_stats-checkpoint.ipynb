```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inflow = pd.read_parquet('/uss/hdsi-prismdata/q1-ucsd-inflows.pqt')
```

```python
# number of transactions per customer
tx_count = (
    inflow.groupby("prism_consumer_id")
    .size()
    .rename("transactions")
)

# Sum of inflow per consumer
inflows = (
    inflow
    .groupby("prism_consumer_id")["amount"]
    .sum()
    .rename("total_inflows")
)

# average inflow amount per consumer
avg_tx_amount = (
    inflow.groupby("prism_consumer_id")["amount"]
    .mean()
    .rename("avg_transaction_amount")
)

#median inflow amount
median_tx_amount = (
    inflow.groupby("prism_consumer_id")["amount"]
    .median()
    .rename("median_transaction_amount")
)

inflow["posted_date"] = pd.to_datetime(inflow["posted_date"], errors="coerce")

# transaction frequency
income_freq = (
    inflow
    .sort_values(["prism_consumer_id", "posted_date"])
    .groupby("prism_consumer_id")["posted_date"]
    .apply(lambda x: x.diff().dt.days.mean())
    .rename("avg_days_between_inflows")
)
```

```python
inflow_stats=pd.concat([tx_count, inflows, avg_tx_amount, median_tx_amount, income_freq], axis=1).reset_index()
```

```python
inflow_stats
```

```python
inflow_stats.describe()
```

```python
numeric_cols = ["transactions", "total_inflows", "avg_transaction_amount", 
                "median_transaction_amount", "avg_days_between_inflows"]

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    plt.hist(inflow_stats[col].dropna(), bins=50, color="skyblue", edgecolor="black")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```
